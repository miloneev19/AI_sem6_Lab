```python
from nltk import sent_tokenize,word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words("english")
```

```python
sent = "I will walk 500 miles and I would walk 500 more , just to be the man who walks a thousand miles to fall down at your door.".lower()
```

```python
sent
```

```python
sentences = sent_tokenize(sent)
print(sentences)
```

```python
words = word_tokenize(sent)
print(words)
```

```python
#for all the words tokenized, transform it into lowercase  -> while iterating over array word=word.lower() or convert it into lower while taking the sent input
```

```python
print(stopwords)
```

```python
for word in words:
  if word in stopwords:
    words.remove(word)
print(words)
```

```python
from nltk import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
print(words)
```

```python
#result will be list hence use square bracket
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
from collections import Counter
from nltk import pos_tag
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
count_dict = {}  #frequency count

for word in stemmed_words:
  if word in count_dict: #using word because it is more prcessed (stemmed)
    count_dict[word]+=1
  else:
      count_dict[word]=1

print(count_dict)
```

```python
#pos tagging  -> converting into tupple so that it becomes immutable
pos_tagged = pos_tag(stemmed_words)
print(pos_tagged)
#nn represents noun , vb refers verb etc
```

```python
count = Counter(tag for _ , tag in pos_tagged)  #everything starting with a captial letter in python is a class here Counter
print(count)
```
